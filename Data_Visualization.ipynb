```python
import matplotlib.pyplot as plt
```

```python
#Lets plot a very basic plot

list1 = [1,2,3,4]

plt.plot(list1)
plt.ylabel('Number Sequence')
plt.show()
```

```python
# To plot X and Y

import math

x =[1,2,3,4]
y = [1 * 2, 2 * math.pow(2,2), 3 * math.pow(2,3), 4 * math.pow(2,4)]

plt.plot(x,y)
plt.show()
```

```python
#Formating your plot

## The default style of your plot is Blue line

plt.plot([1,2,3,4],[1,4,9,16],'ro')

plt.axis((0,6,0,20))
plt.show()
```

```python
import numpy as np

t = np.arange(0,5,0.2)
plt.plot(t, t, 'r--',t,t**2,'bs',t,t**3,'g^')
plt.show()
```

```python
data = {
    'a' : np.arange(50),
    'c' : np.random.randint(0,50,50),
    'd' : np.random.randn(50)
}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

data
```

```python
plt.scatter('a','b', c ='c',s='d', data = data)

plt.xlabel('a values')
plt.ylabel('b values')
plt.show()
```

```python
#Plot a categorical value

# x-axis ->category y-axis -> values

names = ['group a', 'group b', 'group c']
values = [1,10,100]

plt.figure(figsize=(9,3))

plt.subplot(131)
plt.bar(names,values)

plt.subplot(132)
plt.scatter(names,values)

plt.subplot(133)
plt.plot(names, values)

plt.suptitle('Categorical values')

plt.show()
```

```python
import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
iris.head()
```

```python
iris.describe()
```

```python
#Scatter plot of Sepal Length, and Sepal Width

plt.scatter(iris['sepal_length'],iris['sepal_width'])

plt.xlabel('sepal length')
plt.ylabel('sepal width')

plt.title('Scatter plot Sepal Length and Sepal Width')

plt.show()
```

```python
# Box plot

plt.boxplot(x = 'petal_length', data = iris)
plt.title('Box plot of petal length')
plt.show()
```

```python
#Histogram of petal width

plt.hist('petal_width', bins = 10, data =iris)
plt.xlabel('petal width')
plt.ylabel('Frequency')
plt.title("histogram of petal width")
plt.show()
```

```python
#how to create a pie chart

species_count = iris['species'].value_counts()

plt.pie(species_count,labels = species_count.index)
plt.title("Pie chart Distribution of species")

plt.show()
```

```python
sns.set_theme(style='ticks')

sns.pairplot(iris,hue = 'species')
plt.show()
```

```python
sns.set_palette(sns.color_palette("pastel"))

fig,ax = plt.subplots(nrows = 2, ncols = 2, figsize = (15,10))

sns.boxplot(x ='sepal_length', data = iris, hue ='species', ax = ax[0][0])
sns.violinplot(x = 'species', y = 'petal_length', data = iris, ax = ax[0][1])
sns.histplot(x = 'petal_length', bins =10, data =iris, ax =ax[1][0])
sns.kdeplot(x = 'petal_length', hue = 'species',data =iris,ax = ax[1][1])
fig.suptitle("How to make 4 plots at once to understand the information at one go")
plt.show()
```

```python
sns.set_style("whitegrid")

sns.lmplot(x ='petal_length', y = 'sepal_length', data = iris, hue = 'species')
plt.title('LM plot of sepal length and petal length')

plt.show()
```

```python
# Calculate correlation only on numeric columns
# You can select specific numeric columns or drop the non-numeric ones
numeric_iris = iris.drop('species', axis=1) # Drop the 'species' column

# Calculate the correlation matrix
correlation_matrix = numeric_iris.corr()

# Create the heatmap
sns.heatmap(correlation_matrix, annot = True)

plt.show()
```
