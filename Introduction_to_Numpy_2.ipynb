# Array Reshaping--> Reshape, Ravel, Flatten

```python
arr_23 = np.array([1,2,3,4,5,6])
reshaped = arr_23.reshape(2,3)
reshaped
```

```python
reshape2 = reshaped.reshape(3,2)
reshape2
```

```python
# ravel -> convert to 1D array,
ravel = reshape2.ravel()
ravel
```

```python
ravel[0] =100 #changes to ravel will also make changes in reshape2
print(ravel)
print(reshape2)
```

```python
#flatten -> 1 d array.  --> returns the copy of the array
flat = reshape2.flatten()
flat
```

```python
flat[0] =200 #changes to flat will not also make changes in reshape2 although the functionality of ravel and flatten are same
print(flat)
print(reshape2)
```

## Views vs Copies


```python
# Views vs Copies
# view -> original array will be effected
# copies -> original array remains unchanged

arr_30 = np.array([1,2,3,4,5])

view = arr_30[1:3]
view
```

```python
view[0] = 200
```

```python
view
```

```python
arr_30 # updating inn view changed the orginal array
```

```python
#copy

copy = arr_30[1:3].copy()

copy
```
