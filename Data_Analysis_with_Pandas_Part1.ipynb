#Filtering by Conditions

```python
customer
```

```python
#based on the condition : filter customers whose name is 'lilly'

condition = (customer['Name'] == 'lilly')

customer_filter1 = customer[condition]

customer_filter1
```

```python
# multiple conditions : Age>=35 and is female (0: female, 1: male)

condition = (customer['Age'] >= 35 ) & (customer['Gender'] == 0)

customer_filter = customer[condition]
customer_filter
```

```python
#filter the customer name starting with 'L' or 'E'.  use regex in this cases

condition = customer['Name'].str.contains(r'^[E]', regex =True)
customer_filter2 = customer[condition]
customer_filter2
```

```python
name_list = ['lilly','John']

condition = customer['Name'].isin(name_list)

customer[condition]
```

#Replace the missing value



```python
import numpy as np
```

```python
df = pd.DataFrame({
    'accountlength' : [30,20,20,None,30],
    'gender' : ['F','M','F',np.nan,'F'],
    'education' : ['high school','high school','high school','high school',np.nan]

})
df
```

###Comments on NaN
 np.nan is a constant in NumPy representing "Not a Number." It is used to denote missing or undefined values in numerical computations. It is part of the IEEE floating-point standard and is often encountered in datasets where values might be missing or in calculations where results are undefined (e.g., 0/0)

```python
type(np.nan)
```

```python
type(None)
```

```python
df
```


replace the missing value with the average value of the column 'Account length'

```python
value_to_replace = df['accountlength'].mean()

value_to_replace
```

```python
#fill the missing value with value woith value_to_replace
print(df['accountlength'].fillna(value_to_replace) )#this will not replace in the original data frame
print("---------")
df
```

```python

df['accountlength'].fillna(value_to_replace,inplace=True)
df
```

```python
#replace the missing value with the most frequent value in column 'gender'

value_to_replace = df['gender'].mode()[0]

value_to_replace
```

```python
df['gender'].fillna(value_to_replace,inplace=True)
df
```

```python
#assume we know the last customer's education is PHD

df['education'] = df['education'].fillna('phd') #above warnings in future versions ,recommended syntax to avoid bugs
df
```

```python
df = pd.DataFrame({
    'accountlength' : [30,20,20,None,30],
    'gender' : ['F','M','F',np.nan,'F'],
    'education' : ['high school','high school','high school','high school',np.nan]

})
df
```

```python
#assume you know what values you want to fill in place of missing values and fill the all at once with the value known prior

df = df.fillna({'accountlength' : 25,'gender': 'F','education':'phd'})
df
```

#Sorting in the Data **Frame**

```python
data = {
    'Date' : ['2023-09-01','2023-09-02','2023-09-03','2023-09-04'],
    'Product' : ['Product A','Product B','Product A','Product C'],
    'Quantitysold' : [100,150,120,80],
    'Revenue' : [5000,7500,6000,4000]

}
df = pd.DataFrame(data)
df
```

```python
#see the product from highest revenue to the lowest
#df.sort_values(by='Revenue',ascending=False)
#df.sort_values(by='Revenue',ascending=False).reset_index()
df.sort_values(by='Revenue',ascending=False).reset_index(drop=True)
```

```python
#sample data frame

data = {
    'Student_Name' : ['Alice','Bob','Eve','David','Carol'],
    'Age' : [25,30,22,24,28],
    'Score' : [95,93,95,88,88]
}
df = pd.DataFrame(data)
df
```

```python
df = df.sort_values('Score').reset_index(drop=True)
df
```

```python
#sort score from highest to lowest,sort the student age from smallest to highest

df = df.sort_values(by = ['Score','Age'], ascending = [False,True])
df
```
